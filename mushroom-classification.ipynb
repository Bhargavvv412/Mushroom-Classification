### Description
- Desicion Tree is predictive modeling approach. it is used to address datamining and stat and ml problems.
- It has tree structured teminology. so it has a topmost node which is root node which has no incoming edges.
- Terminal node or Leaf node represents a class node and has exactly one incoming node and no outgoing node.
### CART - Classifiacton and Regression Tree
- CART is simple to understand, interpret, visualize and requires little effort for data preparation.
- Regression Tree is mainly use for the target variable is numerical.
### Entropy
- Entropy is measure of misorder or impurity in the given dataset.
-  When the tree finally reaches the terminal or leaf node maximum purity is added.
###  Information Gain
- Information gain is used to decide which feature to split on at each step in building the tree.
- Information gain of a parent node can be calculated as the entropy of the parent node subtracted entropy of the weighted average of the child node.
### Gini Index
- Think of Gini as a “mess meter” that the decision tree uses to keep splitting until baskets are as pure as possible.
### Pruning
- Pruning = cutting useless branches from the tree so it doesn’t get confused and performs better.

Feature Eng

```python
X = df.drop(['class'],axis=1)
y = df['class']
```

```python
X = pd.get_dummies(X)
X.head()
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
y = encoder.fit_transform(y)
print(y)
```

```python
#Splitting the data into train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train.shape, X_test.shape
```

```python
y_train.shape , y_test.shape
```

```python
#Decision Tree Creation
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
from sklearn.metrics import accuracy_score
```

```python
#Using the Decision Tree Classifier with splitting criterion as Gini impurity, the maximum depth of the tree is 3.
clf_gini = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=0)


# fit the model
clf_gini.fit(X_train, y_train)
```

```python
#Plot the tree
plt.figure(figsize=(12,8))

tree.plot_tree(clf_gini.fit(X_train, y_train)) 
```

```python
#Predict the values 
y_pred_gini = clf_gini.predict(X_test)
```

```python
#Predict the value using X train for accuracy comparision 
y_pred_train_gini = clf_gini.predict(X_train)

y_pred_train_gini
```

```python
#Determine the accuracy score
print('Model accuracy score with criterion gini index: {0:0.4f}'. format(accuracy_score(y_test, y_pred_gini)))
#Accuracy Score for training set
print('Training-set accuracy score: {0:0.4f}'. format(accuracy_score(y_train, y_pred_train_gini)))
```
